### Utiliser des index personnalisés

```python
# DataFrame.set_index(Series) =>  par défaut Pandas renvoie un nouveau dataFrame
        # en supprimant la colonne "Series" ayant servi à l'index. Series=nom_colonne
# DataFrame.set_index(Series, inplace=True) inplace=True => modifie le df existant
# DataFrame.set_index(Series, inplace=True, drop=False) drop=False => garde la colonne ayant servi à l'index
fandango_films = fandango.set_index("FILM", drop=False)
fandango_films.index
```

### Sélectionner des valeurs depuis un index personnalisé

```python
# Sélection avec méthode .loc()
fandango_films.loc['Mr. Holmes (2015)']
```

```python
# Sélection avec crochets
fandango_films['Mr. Holmes (2015)': "Two Days, One Night (2014)"]
```

```python
fandango_films.loc['Mr. Holmes (2015)':"Two Days, One Night (2014)"]
```

```python
# Sélection depuis une liste de films
movies = ['Kumiko, The Treasure Hunter (2015)', 'Danny Collins (2015)', 'Ant-Man (2015)']
fandango_films.loc[movies]
```

```python
some_movies = fandango_films.loc[["Top Five (2014)", "Black Sea (2015)", "Taken 3 (2015)"]]
some_movies
```

```python
# Pandas conserve l'ancien Index avec lequel on peut encore sélectionner avec df.iloc[index numérique]
fandango_films.iloc[0]
```

### Méthode apply() sur les colonnes d'un DataFrame

```python
# df.apply(fonction) => applique une fonction sur les colonnes d'un DataFrame
# si c'est une opération vectorielle (ex: écart-type) renvoie une valeur unique
# exepmles
# dataframe.apply(lambda x: x**2)
import numpy as np
types = fandango_films.dtypes
types
```

```python
# filtrer les nombres à virgule (float64) et retourner le nom des colonnes
float_columns = types[types.values == "float64"].index
float_columns
```

```python
# Utilisation des crochets pour filtrer les colonnes de la condition float_columns
float_df = fandango_films[float_columns]
float_df
```

```python
# x ici est un objet Series 
deviations = float_df.apply(lambda x: np.std(x))
deviations
```

```python
types = fandango_films.dtypes
float_columns = types[types.values == "float64"].index
# print(types)
print(float_columns)
float_df = fandango_films[float_columns]
exo_df = float_df.apply(lambda x: x/2)
exo_df[:5]
```

### Méthode apply() sur les lignes d'un DataFrame

```python
# df.apply(function, axis=1)
rt_mt_user = float_df[["RT_user_norm", "Metacritic_user_nom"]]
rt_mt_user[:5]
```

```python
rt_mt_user.apply(lambda x: np.std(x), axis=1)
```

```python
rt_mt_user = float_df[["RT_user_norm", "Metacritic_user_nom"]]
rt_mt_means = rt_mt_user.apply(lambda x: np.mean(x), axis=1)
rt_mt_means[:5]
```
