## Challenge: Analyser de la data

### Introduction au dataset

```python
import pandas as pd
all_ages = pd.read_csv("all-ages.csv")
recent_grads = pd.read_csv("recent-grads.csv")
all_ages.head(5)
```

```python
recent_grads.head(5)
```

### Nombre d'étudiants par catégories de major

```python
# Serie.unique() pour retourner les valeurs uniques d'une Serie
import numpy as np
major_categories = all_ages["Major_category"].unique()
# print(major_categories)

aa_cat_counts = {}
rc_cat_counts = {}
    
for category in major_categories:
    filtre = all_ages[all_ages["Major_category"] == category]
    total = filtre["Total"].sum()
    aa_cat_counts[category] = total

print(aa_cat_counts)

def from_pivot_to_dict(df):
    dico = dict()
    for category in major_categories:
        if not pd.isnull(category):
             dico[category] = df["Total"].loc[category]
    return dico

pre_aa_cat_counts = all_ages.pivot_table(index="Major_category", values="Total", aggfunc=np.sum)
aa_cat_counts = from_pivot_to_dict(pre_aa_cat_counts)
pre_rc_cat_counts = recent_grads.pivot_table(index="Major_category", values="Total", aggfunc=np.sum)
rc_cat_counts = from_pivot_to_dict(pre_rc_cat_counts)
   
print(aa_cat_counts)
print(rc_cat_counts)
```

```python
def total_per_cat(df):
    major_categories = df["Major_category"].unique()
    dictio = {}
    
    for category in major_categories:
        filtre = df[df["Major_category"] == category]
        total = filtre["Total"].sum()
        dictio[category] = total
    return dictio

aa_cat_counts = total_per_cat(all_ages)
print(aa_cat_counts)
rc_cat_counts = total_per_cat(recent_grads)
print(rc_cat_counts)
```
