### Introduction

Explorer des données à l'aide de visualisation graphique

### Présentation du dataset

```python
# pandas.to_datetime()
# df['nom_colonne'] = df.to_datetime(df['nom_colonne'])
import pandas as pd
unrate = pd.read_csv("unrate.csv")
unrate["DATE"] = pd.to_datetime(unrate["DATE"])
unrate[:12]
```

### Introduction à Matplotlib

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.plot()
plt.show()
```

### Ajouter des données à nos graphiques

```python
# plt.plot(x_values, y_values)
# matplotlib accepte des tableaux NumPy ou des objets Series de Pandas

first_year = unrate[:12]
plt.plot(unrate["DATE"][:12], unrate["UNRATE"][:12])
# ou plt.plot(first_year["DATE"], first_year["UNRATE"])
plt.show()
```

### Créer une rotation sur les légendes des données de l'axe x

```python
# plt.xticks()
# matplotlib.pyplot.xticks(*args, **kwargs)
# plt.xticks(rotation=45)
# plt.yticks() pour l'axe des ordonnées
```

```python
first_year = unrate[:12]
plt.plot(first_year["DATE"], first_year["UNRATE"])
plt.xticks(rotation=90)
plt.show()
```

### Ajouter les intitulés des axes et un titre à notre graphique

```python
# plt.xlablel("chaîne")
# plt.ylablel("chaîne")
# plt.title("chaîne")
```

```python
plt.plot(first_year["DATE"], first_year["UNRATE"])
plt.xticks(rotation=90)
plt.xlabel("Mois")
plt.ylabel("Taux de chômage (%)")
plt.title("Evolution du taux de chômage en 1948 aux USA")
plt.show()
```

### Améliorer l'affichage

```python
# fig = plt.figure(figsize(width, height))
```

```python
fig = plt.figure(figsize=(15, 8))
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(first_year["DATE"], first_year["UNRATE"])
ax1.set_title("Taux de chômage en 1948")

second_year = unrate[12:24]
ax2.plot(second_year["DATE"], second_year["UNRATE"])
ax2.set_title("Taux de chômage en 1949")
plt.show()
```

### Comparaison de graphiques

```python
fig = plt.figure(figsize=(12,12))

for i in range(5):
    ax = fig.add_subplot(5, 1, i + 1)
    start_index = i * 12
    end_index = (i + 1) * 12
    subset = unrate[start_index:end_index]
    ax.plot(subset["DATE"], subset["UNRATE"])
    
plt.show()
```

### Superposition de courbes

```python
# pandas.Series.dt
unrate["MONTH"] = unrate["DATE"].dt.month
```

```python
# plt.plot(unrate[0:12]["MONTH"], unrate[0:12]["UNRATE"])
# plt.plot(unrate[12:24]["MONTH"], unrate[12:24]["UNRATE"])
fig = plt.figure(figsize=(6,6))
plt.plot(unrate[0:12]["MONTH"], unrate[0:12]["UNRATE"])
plt.plot(unrate[12:24]["MONTH"], unrate[12:24]["UNRATE"], c="red")
plt.show()
```

```python
fig = plt.figure(figsize=(6,3))
plt.plot(unrate[0:12]["MONTH"], unrate[0:12]["UNRATE"], c="red")
plt.plot(unrate[12:24]["MONTH"], unrate[12:24]["UNRATE"], c="blue")
plt.xlabel("Mois")
plt.ylabel("Taux de chômage (%)")
plt.show()
```

### Ajouter plus de courbes

```python
fig = plt.figure(figsize=(10,6))
colors = ["red", "blue", "green", "orange", "black"]
for i in range(5):
    start_index = i * 12
    end_index = (i + 1) * 12
    subset = unrate[start_index:end_index]
    plt.plot(subset["MONTH"], subset["UNRATE"], c=colors[i])
    
plt.xlabel("Mois")
plt.ylabel("Taux de chômage (%)")
plt.show()
```

### Ajouter une légende

```python
fig = plt.figure(figsize=(10,6))
plt.plot(unrate[:12]['MONTH'], unrate[:12]['UNRATE'], c="red", label='1948')
plt.plot(unrate[12:24]['MONTH'], unrate[12:24]['UNRATE'], c="blue", label='1949')
plt.plot(unrate[24:36]['MONTH'], unrate[24:36]['UNRATE'], c="green", label='1950')
plt.xlabel("Mois")
plt.ylabel("Taux de chômage (%)")
# plt.legend(loc="upper left")
plt.legend(loc="upper left")
plt.show()

# Axes.legend()
```

```python
fig = plt.figure(figsize=(10,6))
colors = ["red", "blue", "green", "orange", "black"]
for i in range(5):
    start_index = i * 12
    end_index = (i + 1) * 12
    subset = unrate[start_index:end_index]
    plt.plot(subset["MONTH"], subset["UNRATE"], c=colors[i], label=1948 + i)
    
plt.xlabel("Mois")
plt.ylabel("Taux de chômage (%)")
plt.legend(loc="upper left")
plt.show()
```

### Courbe finale

```python
# plt.title("chaîne")
```

```python
fig = plt.figure(figsize=(10,6))
colors = ["red", "blue", "green", "orange", "black"]
for i in range(5):
    start_index = i * 12
    end_index = (i + 1) * 12
    subset = unrate[start_index:end_index]
    plt.plot(subset["MONTH"], subset["UNRATE"], c=colors[i], label=1948 + i)
    
plt.xlabel("Mois (de 1 à 12)")
plt.ylabel("Taux de chômage (en %)")
plt.title("Evolution du taux de chômage de 1948 à 1952 aux USA")
plt.legend(loc="upper left")
plt.show()
```
