### Convertir l'âge en une valeur numérique

```python
data['Age'].value_counts()
```

```python
def age_to_num(string):
    if pd.isnull(string):
        return None
    return int(string.split(' ')[0].replace('+', ''))

int_age = data['Age'].apply(age_to_num)
data['int_age'] = int_age
```

```python
data['int_age'].describe()
```

```python
data['How much total combined money did all members of your HOUSEHOLD earn last year?'].value_counts()
```

```python
import re
def money_to_num(string):
    if pd.isnull(string) or re.search('Prefer', string):
        return None
    return int(string.split(' ')[0].replace('$', '').replace(',', ''))
    

int_income = data['How much total combined money did all members of your HOUSEHOLD earn last year?'].apply(money_to_num)
data['int_income'] = int_income
```

```python
data['int_income'].describe()
```

### Corrélation entre distance et revenus

```python
poor_ones = data[data['int_income'] < 150000]
poor_ones["How far will you travel for Thanksgiving?"].value_counts()
```

```python
rich_ones = data[data['int_income'] > 150000]
rich_ones["How far will you travel for Thanksgiving?"].value_counts()
```

```python
rich_ones["How far will you travel for Thanksgiving?"].describe()
```
